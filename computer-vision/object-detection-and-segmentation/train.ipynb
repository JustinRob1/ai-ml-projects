```python
import yaml
import numpy as np
from sklearn.model_selection import train_test_split
from PIL import Image
import os
import matplotlib.pyplot as plt
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define the base path for your files
base_path = '/content/drive/MyDrive/'

# Load your data
train_images = np.load(os.path.join(base_path, 'train_X.npy'))
train_boxes = np.load(os.path.join(base_path, 'train_bboxes.npy'))
train_labels = np.load(os.path.join(base_path, 'train_Y.npy'))
val_labels = np.load(os.path.join(base_path, 'valid_Y.npy'))
val_images = np.load(os.path.join(base_path, 'valid_X.npy'))
val_boxes = np.load(os.path.join(base_path, 'valid_bboxes.npy'))

file_paths = "/content/yolov5"
# Define the directories for your images and labels
train_images_dir = os.path.join(file_paths, 'train/images')
train_labels_dir = os.path.join(file_paths, 'train/labels')
val_images_dir = os.path.join(file_paths, 'val/images')
val_labels_dir = os.path.join(file_paths, 'val/labels')

os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# Iterate over your data
for i in range(train_images.shape[0]):
    # Reshape and normalize the image
    img = train_images[i].squeeze()
    img = img.reshape((64, 64, 3)).astype(np.uint8)

    # Then save the image
    img_obj = Image.fromarray(img)
    img_obj.save(os.path.join(train_images_dir, f'{i}.jpg'))

for i, (image, boxes, labels) in enumerate(zip(train_images, train_boxes, train_labels)):
    with open(os.path.join(train_labels_dir, f'{i}.txt'), 'w') as f:
        for box, label in zip(boxes, labels):
            x_center = (box[1] + box[3]) / 2 / 64
            y_center = (box[0] + box[2]) / 2 / 64
            width = (box[3] - box[1]) / 64
            height = (box[2] - box[0]) / 64
            f.write(f'{label} {x_center} {y_center} {width} {height}\n')

# Iterate over your data
for i in range(val_images.shape[0]):
    # Reshape and normalize the image
    img = val_images[i].squeeze()
    img = img.reshape((64, 64, 3)).astype(np.uint8)

    # Then save the image
    img_obj = Image.fromarray(img)
    img_obj.save(os.path.join(val_images_dir, f'{i}.jpg'))

for i, (image, boxes, labels) in enumerate(zip(val_images, val_boxes, val_labels)):
    with open(os.path.join(val_labels_dir, f'{i}.txt'), 'w') as f:
        for box, label in zip(boxes, labels):
            x_center = (box[1] + box[3]) / 2 / 64
            y_center = (box[0] + box[2]) / 2 / 64
            width = (box[3] - box[1]) / 64
            height = (box[2] - box[0]) / 64
            f.write(f'{label} {x_center} {y_center} {width} {height}\n')

# Create the config file
config = {
    'train': 'train/images',
    'val': 'val/images',
    'nc': 10,  # number of classes
    'names': ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
}

with open(os.path.join(file_paths, 'config.yaml'), 'w') as f:
    yaml.dump(config, f)
```
